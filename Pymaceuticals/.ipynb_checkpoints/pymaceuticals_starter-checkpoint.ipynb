```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
complete_data = pd.merge(mouse_metadata, study_results, how="left", on=["Mouse ID"])

# Display the data table for preview
complete_data
```

```python
# Checking the number of mice.
len(complete_data["Mouse ID"].unique())
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mouse = complete_data[complete_data.duplicated(['Mouse ID', 'Timepoint'])]
len(duplicate_mouse)
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_DataFrame = complete_data.drop_duplicates(subset=['Mouse ID', 'Timepoint'])
clean_DataFrame
```

```python
# Checking the number of mice in the clean DataFrame.
print(len(clean_DataFrame["Mouse ID"].unique()))
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
drug_regimen = clean_DataFrame.groupby(["Drug Regimen"])
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
tumor_mean = drug_regimen["Tumor Volume (mm3)"].mean()
tumor_median = drug_regimen["Tumor Volume (mm3)"].median()
tumor_var = drug_regimen["Tumor Volume (mm3)"].var()
tumor_std = drug_regimen["Tumor Volume (mm3)"].std()
tumor_sem = drug_regimen["Tumor Volume (mm3)"].sem()
# Assemble the resulting series into a single summary DataFrame.
tumor_summary = pd.DataFrame({ "Mean Tumor Volume": tumor_mean, 
                                    "Median Tumor Volume": tumor_median, 
                                    "Variance Tumor Volume": tumor_var, 
                                    "Standard Deviation Tumor Volume": tumor_std,
                                     "SEM Tumor Volume": tumor_sem
                             })
tumor_summary
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
tumor_summary_advanced = drug_regimen["Tumor Volume (mm3)"].agg(["mean", "median", "var", "std", "sem"])
# Using the aggregation method, produce the same summary statistics in a single line
tumor_summary_advanced = tumor_summary_advanced.rename(columns={"mean": "Mean Tumor Volume",
                                                  "median": "Median Tumor Volume",
                                                   "var": "Variance Tumor Volume",
                                                   "std":"Standard Dev Tumor Volume ", 
                                                   "sem":"SEM Tumor Volume"})

tumor_summary_advanced
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
final_vol = clean_DataFrame.groupby(["Mouse ID"])["Timepoint"].max()
final_vol

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
final_volmerged = pd.merge(final_vol, clean_DataFrame, on=(["Mouse ID", "Timepoint"]))
final_volmerged
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for treatment in treatments: 
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumor_volume = final_volmerged.loc[final_volmerged["Drug Regimen"] == treatment, "Tumor Volume (mm3)"]
    # add subset 
    tumor_vol_data.append(tumor_volume)
    
    
    # Determine outliers using upper and lower bounds
    quartiles = tumor_volume.quantile([.25, .5, .75])
    lowerq = quartiles[.25]
    upperq = quartiles[.75]
    iqr = upperq - lowerq 
    lower_key = lowerq - (1.5*iqr)
    upper_key = upperq + (1.5*iqr)
    
    print(f"For {treatment} values below {lower_key} and above {upper_key} are outliers.")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
fig1,ax1 = plt.subplots(figsize=(12,7))
flierprops = dict(marker='+', markerfacecolor='g', markersize=15,
                  linestyle='none', markeredgecolor='r')
ax1.boxplot(tumor_vol_data, flierprops=flierprops)
ax1.set_title('Tumor Volume Per Drug Treatment')
ax1.set_ylabel('Final Tumor Volume (mm3)')
ax1.set_xlabel('Drug Regimen')
plt.xticks([1, 2, 3, 4], ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin'])
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
Capomulin_treatment = clean_DataFrame.loc[clean_DataFrame["Drug Regimen"] == "Capomulin"]
Capomulin_treatment.value_counts("Mouse ID")
t565 = clean_DataFrame.loc[clean_DataFrame["Mouse ID"] == "t565"]
t565
fig1,ax1 = plt.subplots(figsize=(12,7))
cap_xaxis = t565["Timepoint"]
cap_yaxis = t565["Tumor Volume (mm3)"]
plt.plot(cap_xaxis, cap_yaxis)
plt.xlabel("Timepoint (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Tumor Volume Over Time for Mouse t565")
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
avg_cap_volume = Capomulin_treatment.groupby(["Mouse ID"]).mean()
plt.scatter(avg_cap_volume['Weight (g)'], avg_cap_volume['Tumor Volume (mm3)'])
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Average Tumor Volume vs Weight Using Capomulin Treatment")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
corr_coef = st.pearsonr(avg_cap_volume['Weight (g)'], avg_cap_volume['Tumor Volume (mm3)'])[0]
print(f"The correlation between mouse weight and average tumor volume is {corr_coef}")

plt.scatter(avg_cap_volume['Weight (g)'], avg_cap_volume['Tumor Volume (mm3)'])
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")  # Fixed the ylabel here

lin_x_values = avg_cap_volume['Weight (g)']
lin_y_values = avg_cap_volume['Tumor Volume (mm3)']

# Linear regression
(slope, intercept, rvalue, pvalue, stderr) = linregress(lin_x_values, lin_y_values)
regress_values = lin_x_values * slope + intercept
equation = "y = " + str(round(slope, 2)) + "x + " + str(round(intercept, 2))
print(equation)

# Add the line to the plot
plt.annotate(equation, xy=(20, 40), xycoords="data", xytext=(0.2, 0.80), textcoords="axes fraction", fontsize=12, color="red")
plt.plot(lin_x_values, regress_values, "r-")
plt.title("Average Tumor Volume vs Weight Using Capomulin Treatment")
plt.show()
```
